```python
df=pd.read_csv("/content/ford_car_dataset.csv")
```

```python
df
```

```python
print("Independent Features:")
X=df.drop("price",axis=1)
X.head()
```

```python
print("Dependent Features:")
Y=df["price"]
Y.head()
```

```python
print("Shape of X:",X.shape)
```

```python
print("Shape of Y:",Y.shape)
```

```python
df_encoded=pd.get_dummies(X,columns=["model","transmission","fuelType"])
df_encoded=df_encoded.astype(int)
print("After Encoding:")
print(df_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

col=["year","mileage","tax","mpg","engineSize"]

scaler=StandardScaler()

df_encoded[col]=scaler.fit_transform(df_encoded[col])
df_encoded[col].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train,X_test,Y_train,Y_test=train_test_split(df_encoded,Y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(X_train,Y_train)
```

```python
print("Intercept:",model.intercept_)
print("\nCoefficients:",model.coef_)
```

```python
y_pred=model.predict(X_test)
print("Predicted value:",y_pred[:10])
print("\nActual value:",Y_test.tolist()[:10])
```

```python
from sklearn.metrics import r2_score

r2=r2_score(Y_test,y_pred)
print("R2 Score:",r2)
```

```python
#R2 is 84 means the model explains 84% of the variation in car prices.
#It is performing very well
```
